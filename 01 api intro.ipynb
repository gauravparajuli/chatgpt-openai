```python
import openai
import tiktoken
from google.colab import userdata

api_key = userdata.get('OPENAI_API_KEY')

client = openai.Client(api_key=api_key)
```

# helper function

```python
def get_completion(prompt, model='gpt-3.5-turbo'):
  messages = [{'role': 'user', 'content': prompt}]
  response = client.chat.completions.create(
      model=model,
      messages=messages,
      temperature=1
  )

  return response.choices[0].message.content
```

```python
get_completion('What is the capital of nepal')
```

```python
get_completion('reverse letters in NEPAL')
```

```python
get_completion('reverse letters in N-E-P-A-L')
```

# helper function in chat format

```python
def get_completion_from_messages(messages, model='gpt-3.5-turbo', temperature=0, max_tokens=500):
  response = openai.chat.completions.create(
      model=model,
      messages=messages,
      temperature=temperature, # degree of randomness
      max_tokens=max_tokens # max number of tokens that model can output
  )

  return response.choices[0].message.content
```

```python
messages = [
    dict(role='system', content='you are an assitant who responds in the style of dr. seuss'),
    dict(role='user', content='write me a very short poem about a happy child')
]
print(get_completion_from_messages(messages, temperature=1))
```

```python
def get_completion_and_token_count(messages, model='gpt-3.5-turbo', temperature=0, max_tokens=500):
  response = openai.chat.completions.create(
      model=model,
      messages=messages,
      temperature=temperature, # degree of randomness
      max_tokens=max_tokens # max number of tokens that model can output
  )

  content = response.choices[0].message.content

  token_dict = dict(prompt_tokens=response.usage.prompt_tokens, completion_tokens=response.usage.completion_tokens, total_tokens=response.usage.total_tokens)

  return content, token_dict

get_completion_and_token_count(messages)
```
